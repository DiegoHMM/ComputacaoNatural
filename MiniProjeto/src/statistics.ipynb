```python
import pandas as pd
from indiv import *

seeds = [42,19,34,84,91]
```

```python
df_statistics = []
```

```python
for seed in seeds:
    df_statistics.append(pd.read_csv('Statistics/statistics_' + str(seed) + '.csv'))
```

```python
# Agrupando por índice e calculando a média
df_concat = pd.concat(df_statistics)
df_mean = df_concat.groupby(df_concat.index).mean()
```

```python
df_mean.head(5)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
# Definindo um estilo para os gráficos
sns.set(style='whitegrid')

plt.figure(figsize=(8, 5))  # Define o tamanho do gráfico

# Plota a linha para 'best_fitness'
sns.lineplot(x=df_mean['gen'], y=df_mean['best_fitness'], label='Melhor Fitness', color='orange')

# Plota a linha para 'mean_fitness'
sns.lineplot(x=df_mean['gen'], y=df_mean['mean_fitness'], label='Média Fitness', color='blue')

plt.xlabel('Geração')  # Define o rótulo do eixo x
plt.ylabel('Fitness')  # Define o rótulo do eixo y
plt.title('Geração X Fitness')  # Define o título do gráfico
plt.legend()  # Mostra a legenda
plt.show()  # Mostra o gráfico
```
